## Linear Regression

### Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd

dataset = pd.read_csv('Employee Attrition.csv')
#variable and this is a function for uploading the dataset
```

### Rearranging Columns

```python
dataset_reordered = dataset[dataset.columns[:1].tolist() + dataset.columns[2:].tolist() + [dataset.columns[1]]]

dataset_reordered.to_csv('reordered_employee_attrition.csv', index=False)

dataset_reordered.head() 
```

### Checking Data Types

```python
dataset_reordered.dtypes
```

### Converting categorical data using label encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder
```

```python
dataset_reordered['salary'] = label_encoder.fit_transform(dataset_reordered['salary'])
dataset_reordered.head()
```

### Converting Data Types(integer to float)

```python
import pandas as pd

# Convert integer columns to float
int_columns = ['salary']
dataset_reordered[int_columns] = dataset_reordered[int_columns].astype(float)

# Verify the conversion
print(dataset_reordered.dtypes)
```

### Converting Categorical Data using One-Hot Encoding

```python

# Perform one-hot encoding on all categorical columns
dataset_encoded = pd.get_dummies(dataset_reordered, drop_first=True)

# Display the first few rows of the transformed dataset
dataset_encoded.head()
```

```python
dataset_encoded.dtypes
```

### Converting Data Types (bool to float)

```python
# Convert the bool to float
dataset_encoded['dept_RandD'] = dataset_encoded['dept_RandD'].astype(float)
dataset_encoded['dept_accounting'] = dataset_encoded['dept_accounting'].astype(float)
dataset_encoded['dept_hr'] = dataset_encoded['dept_hr'].astype(float)
dataset_encoded['dept_management'] = dataset_encoded['dept_management'].astype(float)
dataset_encoded['dept_marketing'] = dataset_encoded['dept_marketing'].astype(float)
dataset_encoded['dept_product_mng'] = dataset_encoded['dept_product_mng'].astype(float)
dataset_encoded['dept_sales'] = dataset_encoded['dept_sales'].astype(float)
dataset_encoded['dept_support'] = dataset_encoded['dept_support'].astype(float)
dataset_encoded['dept_technical'] = dataset_encoded['dept_technical'].astype(float)

# Check the data types
print(dataset_encoded.dtypes)
```

### Rearranging Columns

```python
# Move the target column to the end
target_col = 'satisfaction_level'  # Replace with your actual target column name
dataset_encoded = dataset_encoded[[col for col in dataset_encoded.columns if col != target_col] + [target_col]]

# Display the modified DataFrame
dataset_encoded.head()
```

### Checking for NaN values

```python
dataset_encoded.isna()
dataset_encoded.isnull()
print(dataset_encoded.isna().sum())
# This will show the number of NaN values in each column.
```

### Handling Missing Values

```python
# Assign the filled values directly back to the DataFrame columns
dataset_encoded['Emp ID'] = dataset_encoded['Emp ID'].fillna(dataset_encoded['Emp ID'].mean())
dataset_encoded['last_evaluation'] = dataset_encoded['last_evaluation'].fillna(dataset_encoded['last_evaluation'].mean())
dataset_encoded['number_project'] = dataset_encoded['number_project'].fillna(dataset_encoded['number_project'].mean())
dataset_encoded['average_montly_hours'] = dataset_encoded['average_montly_hours'].fillna(dataset_encoded['average_montly_hours'].mean())
dataset_encoded['time_spend_company'] = dataset_encoded['time_spend_company'].fillna(dataset_encoded['time_spend_company'].mean())
dataset_encoded['Work_accident'] = dataset_encoded['Work_accident'].fillna(dataset_encoded['Work_accident'].mean())
dataset_encoded['promotion_last_5years'] = dataset_encoded['promotion_last_5years'].fillna(dataset_encoded['promotion_last_5years'].mean())
dataset_encoded['satisfaction_level'] = dataset_encoded['satisfaction_level'].fillna(dataset_encoded['satisfaction_level'].mean())

print(dataset_encoded.isna().sum()) 
```

### Checking for Duplicates

```python
duplicates = dataset_encoded.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

### Removing Duplicates

```python
# Remove duplicates
dataset_cleaned = dataset_encoded.drop_duplicates(keep='first')

# Verify that duplicates have been removed
remaining_duplicates = dataset_cleaned.duplicated().sum()
print(f"Remaining duplicate rows: {remaining_duplicates}")
```

### Getting the inputs and output

```python
# [rows,columns]
X = dataset_cleaned.iloc[:,1:-1].values
X
```

```python
y = dataset_cleaned.iloc[:,-1].values
y
```

### Creating the Training Set and the Test Set

```python
# scikitlearn is a library
# model_selection is a module
# train_test_split is a function
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2,random_state=0)
```

```python
# fit is a method inside LinearRegression class - they are like functions.
model.fit(X_train, y_train)
```

## Part 3: Evaluating the Model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model training with Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train_scaled, y_train)

# Predictions
y_pred = model.predict(X_test_scaled)

```

### R-Squared

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```

### Adjusted R-Squared

```python
k = X_test.shape[1]
n = X_test.shape[0]
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
adj_r2
```

```python
from sklearn.metrics import mean_squared_error


# Calculate the Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (Min-Max Scaled): {mse}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np


# Perform cross-validation
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')

# Calculate mean and standard deviation of MSE
mean_cv_mse = -cv_scores.mean()  # Negate to get MSE
std_cv_mse = cv_scores.std()      # Standard deviation of MSE

print(f'Mean Cross-Validated MSE: {mean_cv_mse}')
print(f'Standard Deviation of MSE: {std_cv_mse}')
```

### Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(0)  # For reproducibility
X = 2 * np.random.rand(100, 1)  # 100 random values for X
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relationship with some noise

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
predicted = model.predict(X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='pink', label='Data Points')  # Scatter plot of actual data
plt.plot(X, predicted, color='blue', label='Regression Line')  # Regression line
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
